```python
import pandas as pd
import os
```

```python
csv_file = "nuforc_reports.csv"
ufo_df = pd.read_csv(csv_file)
ufo_df.head(20)
```

```python
ufo_df = ufo_df.rename(columns={'city_latitude': 'latitude', 'city_longitude': 'longitude'})
ufo_df.head(20)
```

```python
ufo_df.to_csv("ufos_cleaned.csv", index=False)
```

```python
csv_file2 = "scrubbed.csv"
ufo_scrubbed_df = pd.read_csv(csv_file2)
ufo_scrubbed_df.head(20)
```

```python
ufo_scrubbed_df = ufo_scrubbed_df.dropna()
ufo_scrubbed_df = ufo_scrubbed_df.drop(columns='country')
ufo_scrubbed_df = ufo_scrubbed_df.rename(columns={'duration (seconds)': 'seconds', 'duration (hours/min)': 'hours_min', 'date posted': 'date_posted'})
ufo_scrubbed_df.head(20)
```

```python
ufo_scrubbed_df.count()
```

```python
ufo_scrubbed_df.to_csv("ufo_scrubbed.csv", index=False)
```
