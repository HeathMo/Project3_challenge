```python
import pandas as pd
import os
import requests
import json
import urllib
```

```python
csv_file2 = "scrubbed.csv"
ufo_scrubbed_df = pd.read_csv(csv_file2)
ufo_scrubbed_df.head(20)
```

### Dropping columns not needed for analysis

```python
ufo_scrubbed_df = ufo_scrubbed_df.drop(columns=['duration (hours/min)', 'date posted'])
ufo_scrubbed_df.head(20)
```

### Perform a count to see how much data is incomplete

```python
ufo_scrubbed_df.count()
```

### Drop incomplete data points as not required for visualizations

```python
ufo_scrubbed_df = ufo_scrubbed_df.dropna()
ufo_scrubbed_df.head(10)
```

### Perform count to see if drop consolidated the data

```python
ufo_scrubbed_df.count()
```

### Reset index

```python
ufo_scrubbed_df = ufo_scrubbed_df.reset_index(drop=True)
ufo_scrubbed_df.head(10)
```

```python
ufo_scrubbed_df.to_csv("ufo_quick_scrubbed.csv", index=False)
```
